```python
import numpy as np
import matplotlib.pyplot as plt

p = 0.6
n = 100000
samples = np.random.geometric(p, size=n)

# Histograma normalizado
max_k = np.max(samples)
bins = np.arange(1, max_k + 2) - 0.5
hist_vals, _ = np.histogram(samples, bins=bins)
hist_vals_normalized = hist_vals / n

# PMF usando la fórmula
k_vals = np.arange(1, max_k + 1)
pmf_vals = p * (1 - p) ** (k_vals - 1)  # Fórmula directa

# Gráfica
plt.figure(figsize=(10, 6))
plt.bar(k_vals, hist_vals_normalized, width=1, alpha=0.6, label="Histograma (estimado)")
plt.plot(k_vals, pmf_vals, 'ro-', label="PMF fórmula")
plt.xlabel("Número de ensayos hasta el éxito")
plt.ylabel("Probabilidad")
plt.title("Distribución Geométrica (p=0.3)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --------------------------
# 1. Parámetro de la distribución
# --------------------------
p = 0.3  # Probabilidad de éxito en cada ensayo

# --------------------------
# 2. Generar n muestras de la distribución geométrica
# --------------------------
n = 100000 # Número de simulaciones
samples = np.random.geometric(p, size=n)  # Cuenta ensayos hasta el primer éxito (comienza en 1)

# --------------------------
# 3. Crear histograma de frecuencias relativas (estimación de PMF)
# --------------------------
max_k = np.max(samples)  # Para saber hasta qué valor graficar
bins = np.arange(1, max_k + 2) - 0.5  # Definimos bins centrados en los enteros
hist_vals, _ = np.histogram(samples, bins=bins)
hist_vals_normalized = hist_vals / n  # Normalizar para obtener probabilidades

# --------------------------
# 4. Calcular PMF teórica usando la fórmula explícita
# --------------------------
k_vals = np.arange(1, max_k + 1)  # Valores posibles de la variable
pmf_teorica = p * (1 - p) ** (k_vals - 1)  # PMF: p(1-p)^(x-1)

# --------------------------
# 5. Calcular media y varianza teóricas
# --------------------------
media = 1 / p
varianza = (1 - p) / p**2

# --------------------------
# 6. Graficar el histograma y la PMF teórica
# --------------------------
plt.figure(figsize=(10, 6))

# Histograma (estimación de la PMF)
plt.bar(k_vals, hist_vals_normalized, width=1, alpha=0.6, label="Histograma (estimado)")

# PMF teórica con la fórmula
plt.plot(k_vals, pmf_teorica, 'ro-', label="PMF teórica (fórmula)")

# Línea vertical para la media
plt.axvline(media, color='g', linestyle='--', linewidth=2, label=f"Media = {media:.2f}")

# Línea vertical para ±1 desviación estándar
std = np.sqrt(varianza)

# --------------------------
# 7. Decoración del gráfico
# --------------------------
plt.xlabel("Número de ensayos hasta el primer éxito")
plt.ylabel("Probabilidad")
plt.title(f"Distribución Geométrica (p = {p})")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
